```python
import SimpleITK as sitk
import matplotlib.pyplot as plt


def denoising(img):
    # Use MedianImageFilter instead of BilateralImageFilter
    median_filter = sitk.MedianImageFilter()
    denoised_image = median_filter.Execute(img)
    return denoised_image
```

```python
imag = sitk.ReadImage("D:/fin_de_etude/data/subset1/1.3.6.1.4.1.14519.5.2.1.6279.6001.106719103982792863757268101375.mhd")
imagd = denoising(imag)
fig, ax = plt.subplots(1,2, figsize=(10, 15))
imgd = sitk.GetArrayFromImage(imagd)
img = sitk.GetArrayFromImage(imag)
ax[0].imshow(img[int(img.shape[0]/2), :, :], cmap='gray')
ax[0].set_title('Original Image')

# Display the denoised image on the right subplot
ax[1].imshow(imgd[int(imgd.shape[0]/2), :, :], cmap='gray')

ax[1].set_title('Denoised Image')


# Show the plots
plt.show()
```
