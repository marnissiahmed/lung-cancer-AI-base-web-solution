```python
import tensorflow as tf
import pandas as ad
import os 
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
```

```python
path = "D:/fin_de_etude/data/data2D/nodule/"
deeplab = tf.keras.models.load_model('D:/fin_de_etude/data/pfe/test/deeplabv3plus.h5', compile=False)
for file in os.listdir(path):
        img = np.load(os.path.join(path,file))
        img = img/255
        out = deeplab.predict(img.reshape(1,512,512,1))
        out=np.where(out>0.5,1,0)
        np.save(f"D:/fin_de_etude/data/classification data/nodule/{file}.npy",out)
```

```python
path = "D:/fin_de_etude/data/data2D/nonnodule/"
deeplab = tf.keras.models.load_model('D:/fin_de_etude/data/pfe/test/deeplabv3plus.h5', compile=False)
for file in os.listdir(path):
        img = np.load(os.path.join(path,file))
        img = img/255
        out = deeplab.predict(img.reshape(1,512,512,1))
        out=np.where(out>0.5,1,0)
        np.save(f"D:/fin_de_etude/data/classification data/nonnodule/{file}.npy",out)
```

```python
def get_tensor(path):
    images_liste = os.listdir(path)
    Nb_images = len(images_liste)
    tenseur = np.empty((Nb_images, 512, 512,1),dtype="uint8")

    for i,file in enumerate(images_liste):
      
        img = np.load(os.path.join(path,file))
        tenseur[i, :, :, :] = img
    return tenseur
```

```python
X0 = get_tensor("D:/fin_de_etude/data/classification data/nonnodule/")
X1 = get_tensor("D:/fin_de_etude/data/classification data/nodule/")
```
